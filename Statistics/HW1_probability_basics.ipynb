### Задача 1
Из колоды в $52$ карты вынимают случайным образом $4$ карты. Найти число исходов, соответствующих тому, что был вытянут хотя бы один туз.

### Решение
В условии не сказано, упорядоченная или неупорядоченная выборка. Посчитаем для неупорядоченной. Если порядок выемки карт имеет значение, то исходов будет просто больше в 4! раз. Это количество перестановок 4 карт. 

Число исходов для случая, когда порядок выемки карт не важен, равно сумме числа исходов, когда мы вынули ровно один туз, ровно два туза, ровно три туза и ровно 4 туза.<br>
В каждом варианте это число сочетаний из 4 тузов необходимого количества тузов + число сочетаний оставшегося числа карт из не тузов.<br>
Имеем:

$$\displaystyle C_4^1 \cdot C_{48}^{3} + C_4^2 \cdot C_{48}^2 + C_4^3 \cdot C_{48}^1 + C_4^4 = 76145$$ 

### Задача 2
Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что рядом будут сидеть:

а) два определённых человека?

б) три определённых человека?

### Решение<br>
Всего вариантов рассадки 7! (число перестановок).

а) Чтобы посчитать число рассадок, при которых 2 человека сидят рядом, представим этих двух людей единым целым. Нужно рассадить 5 человек и группу из двух. То есть 6 всего. Число рассадок $6!$. Еще нужно учесть, что внутри группы люди тоже могут меняться местами, оставшись сидеть рядом. Таких перестановок 2.

Итого ответ: $\displaystyle \frac{6! \cdot 2!}{7!} = \frac{2}{7}$

б) Аналогично считаем. Сначала рассаживаем 4 человека плюс группу из трех и затем учитываем пересадки среди троих.

Итого: $\displaystyle \frac{5! \cdot 3!}{7!} = \frac{1}{7}$

### Задача 3
Из $60$ вопросов, входящих в экзаменационные билеты, студент знает $50$. Какова вероятность того, что среди трёх наугад выбранных вопросов студент знает:

а) все?

б) два?

### Решение
Элементарными исходами будет неупорядоченный набор выбранных трех билетов. <br>Всего элементарных исходов $C_{60}^3$

а) Исходов, при которых студент выбирает 3 билета, которые студент знает, равна $C_{50}^3$.<br>
Вероятность того, что студент знает все три выбранных билета

$$\displaystyle \frac {C_{50}^3}{C_{60}^3} = \frac{50 \cdot 49 \cdot 48}{60 \cdot 59 \cdot 58} = \frac{980}{1711} \approx 0.57$$

б) Исходов, при которых студент выбирает 2 хороших билета билета равна $\displaystyle C_{50}^2 \cdot C_{10}^1$.<br>
Вероятность того, что студент лишь два выбранных билета

$$\displaystyle \displaystyle \frac {C_{50}^2 \cdot C_{10}^1}{C_{60}^3} = \frac{50 \cdot 49 \cdot 3 \cdot 10}{60 \cdot 59 \cdot 58} = \frac{1225}{3422} \approx 0.36$$

### Задача 4
Бросается игральная кость. Пусть событие $A$ - появление чётного числа, событие $B$ - появление числа больше трёх. Являются ли эти события независимыми?

### Решение
$\displaystyle P(A\cap B) = \frac{1}{3}$

$\displaystyle P(A) = \frac{1}{2}$

$\displaystyle P(B) = \frac{1}{2}$

$\displaystyle P(A \cap B) \neq P(A)\cdot P(B) = \frac{1}{4}$

События не являются независимыми.

### Задача 5
Допустим, имеется некоторая очень редкая болезнь (поражает $0.1\%$ населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет $99\%$ больных этой болезнью и всего лишь в $1\%$ случаев даёт ложный положительный ответ.<br>
Вопрос: какова вероятность, что вы действительно больны ей?

### Решение
$A$ – тест положительный<br>
$B$ – вы больны<br>
$\overline{B}$ – вы больны<br>
$B|A$ - вы действительно больны при условии, что тест положительный<br>
$A|B$ - тест положительный при условии, что вы больны<br>
$A|\overline{B}$ - тест ошиюочно положительный при условии, что вы здоровы<br>
$P(A|B) = 99\%$<br>
$P(A|\overline{B}) = 1\%$<br>
$P(B) = 0.1\%$<br>
$P(\overline{B}) = 99.9\%$<br>
$\displaystyle P(B|A) = \frac{P(A|B)P(B)}{P(A)} = \frac{P(A|B)P(B)}{P(A|B)P(B) + P(A|\overline{B})P(\overline{B})} = \frac{99\% \cdot 0.1\%}{99\% \cdot 0.1\% + 1\% \cdot 99.9\%} = \frac{9.9}{9.9 + 99.9} = \frac{11}{122} \approx 9\%$
