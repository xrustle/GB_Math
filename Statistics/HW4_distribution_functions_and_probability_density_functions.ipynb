# Непрерывные случайные величины. Функция распределения и плотность распределения вероятностей. Равномерное и нормальное распределение. Центральная предельная теорема
### Задача 1

О непрерывной равномерно распределённой случайной величине `B` известно, что её дисперсия равна 0.2. Можно ли найти правую границу величины `B` и её среднее значение, зная, что левая граница равна 0.5? Если да, найдите их.

#### Решение
Математическое ожидание и дисперсия равномерного распределения:

$M(B) = \dfrac{a + b}{2}, \:\: D(B) = \dfrac{(b - a)^2}{12}.$

Из условия $a = 0.5, \:\: D(B) = 0.2$

Нам нужно найти $M(B)$ и $b$.

У нас два уравнения и две неизвестных. Есть шанс найти неизвестные.

Найдем a из уравнения для дисперсии:

$0.2 = \dfrac{(b - 0.5)^2}{12} \:\: \Leftrightarrow \:\: (b - 0.5)^2 = 2.4 \:\: \Leftrightarrow \:\: b_{1,2} = \pm\sqrt{2.4} + 0.5 $

Учтем уcловие, что $b > a$.

Значит, $b = \sqrt{2.4} + 0.5 \approx 2.05$

$M(B) = \dfrac{a + b}{2} = \dfrac{1}{2}\left(\sqrt{2.4} + 1\right) \approx 1.27$

### Задача 2

На сколько сигм (средних квадратических отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой $M(X) = 178$ и $D(X) = 25$?

#### Решение
Будем считать, что случайная величина, равная росту человека имеет нормальное распределение.

Тогда $\sigma = \sqrt{D(X)} = 5$

```python
(190 - 178) / 5
```

Получаем, что 190 см отклоняется от математического ожидания на 2.4 сигм.

### Задача 3

Рост взрослого населения города `X` имеет нормальное распределение, причём, средний рост равен 174 см, а среднее квадратическое отклонение равно 8 см. С помощью правил стандартного отклонения, двух и трёх сигм посчитайте, какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

* больше 182 см?
* больше 190 см?
* от 166 см до 190 см?
* от 166 см до 182 см?
* от 158 см до 190 см?
* не выше 150 см или не ниже 190 см?
* не выше 150 см или не ниже 198 см?
* ниже 166 см?

#### Решение

$a + 3\sigma = 198$

$a + 2\sigma = 190$

$a + \sigma = 182$

$a - \sigma = 166$

$a - 2\sigma = 158$

$a - 3\sigma = 150$

Зная, следующие правила:

* Интервал от $a - \sigma$ до $a + \sigma$ (__стандартное отклонение__) содержит около $68\%$ вероятностной массы,
* От $a - 2\sigma$ до $a + 2\sigma$ — около $95\%$ массы (__правило двух сигм__),
* От $a - 3\sigma$ до $a + 3\sigma$ — около $99.7\%$ массы (__правило трёх сигм__),

посчитаем необходимые вероятности.

* больше 182 см? 

$\approx (100\% - 68\%) / 2 = 16\%$
* больше 190 см? 

$\approx (100\% - 95\%) / 2 = 2.5\%$
* от 166 см до 190 см? 

$\approx (68\% + 95\%) / 2 = 81.5\%$
* от 166 см до 182 см? 

$\approx 68\%$
* от 158 см до 190 см?

$\approx 95\%$
* не выше 150 см или не ниже 190 см? 

$\approx (100\% - 99.7\%) / 2 + (100\% - 95\%) / 2 = 2.65\%$
* не выше 150 см или не ниже 198 см? 

$\approx 100\% - 99.7\% = 0.3\%$
* ниже 166 см? 

$\approx (100\% - 68\%) / 2 = 16\%$
