__1.__ Решить систему уравнений методом Гаусса:

$\left\{\begin{array}{l}
{x_{1}+x_{2}-x_{3}-2x_{4}=0} \\
{2x_{1}+x_{2}-x_{3}+x_{4}=-2} \\
{x_{1}+x_{2}-3x_{3}+x_{4}=4}
\end{array}\right.$

#### Решение

Запишем расширенную матрицу системы:

$\left(
\begin{array}{cccc}
{1} & {1} & {-1} & {-2}\\ 
{2} & {1} & {-1} & {1}\\ 
{1} & {1} & {-3} & {1}
\end{array}\right|$
$\left.\begin{array}{c}
{0}\\ 
{-2}\\
{4}
\end{array}\right)$

Путем элементарных преобразований приведем матрицу к ступенчатому виду.

Вычтем из второй строки первую, умноженную на 2, и из третьей строки вычтем первую:

$\left(
\begin{array}{cccc}
{1} & {1} & {-1} & {-2}\\ 
{0} & {-1} & {1} & {5}\\ 
{0} & {0} & {-2} & {3}
\end{array}\right|$
$\left.\begin{array}{c}
{0}\\ 
{-2}\\
{4}
\end{array}\right)$

Оставшаяся матрица соответствует системе

$\left\{\begin{array}{l}
{x_{1}+x_{2}-x_{3}-2{x_4}=0} \\
{-x_{2}+x_{3} +5 x_4=-2} \\
{-2 x_3 + 3 x_4=4}
\end{array}\right.$

В качестве свободного параметра возьмем $x_4 = c$.

Тогда из последнего уравнения $x_3 = -2 + 1.5 c$

Из второго уравнения $x_2 = 2 + x_3 + 5 x_4 = 1.5c + 5c = 6.5c$

И, наконец, $x_1 = -x_2 + x_3 + 2x_4 = -6.5c -2 +1.5c+2c = -3c -2$

#### Ответ: 
Общее решение $\left\{\begin{array}{l}
{x_1 = -3c -2}\\
{x_2 = 6.5c}\\
{x_3 = -2 + 1.5c}\\
{x_4 = c}
\end{array}\right.$

__4.__ Решить систему линейных уравнений методом Холецкого

$\left\{\begin{array}{l}
81x_{1}-45x_{2}+45x_{3}=531 \\
-45x_{1}+50x_{2}-15x_{3}=-460 \\
45x_{1}-15x_{2}+38x_{3}=193
\end{array}\right.$

#### Решение

$A = \left(\begin{array}{ccc}
{81} & {-45} & {45} \\ 
{-45} & {50} & {-15} \\ 
{45} & {-15} & {38}
\end{array}\right)$

$Ax=LL^{T}x=Ly=b$

Ищем матрицу $L$ по формулам.

$l_{ii}=\sqrt{a_{ii}-\sum_{k=1}^{i-1}l_{ik}^{2}},$

$l_{ij}=\frac{1}{l_{jj}}\left( a_{ij}-\sum_{k=1}^{j-1}l_{ik}l_{jk}\right ), \; j < i.$

Произведем разложение на $LL^{T}$:

$l_{11}=\sqrt{a_{11}}=\sqrt{81} = 9$

$l_{21}=\dfrac{a_{21}}{l_{11}}=-\dfrac{45}{9} = -5$

$l_{31}=\dfrac{a_{31}}{l_{11}}=\dfrac{45}{9} = 5$

$l_{22}=\sqrt{a_{22}-l_{21}^{2}}=\sqrt{50 - 25}= 5$

$l_{32}=\frac{1}{l_{22}}\left ( a_{32}-l_{21}l_{31} \right)=\dfrac{-15 + 5\cdot 5}{5}= 2$

$l_{33}=\sqrt{a_{33}-l_{13}^{2}-l_{23}^{2}}=\sqrt{38-25-4}=3$

Получили матрицы:

$L = \left(\begin{array}{ccc}
{9} & {0} & {0} \\ 
{-5} & {5} & {0} \\ 
{5} & {2} & {3}
\end{array}\right)$

$L^{T} = \left(\begin{array}{ccc}
{9} & {-5} & {5} \\ 
{0} & {5} & {2} \\ 
{0} & {0} & {3}
\end{array}\right)$

Сначала решаем СЛАУ $Ly=b$

$\left\{\begin{array}{l}
9y_{1}=531 \\
-5y_{1}+5y_{2}=-460 \\
5y_{1}+2y_{2}+3y_{3}=193
\end{array}\right.$

$\left\{\begin{array}{l}
y_{1}=59 \\
y_{2}=\dfrac{-460 + 5\cdot 59}{5} = -33\\
y_{3}=\dfrac{193 + 2 \cdot 33 - 5 \cdot 59}{3} = -12
\end{array}\right.$

Сначала решаем СЛАУ $L^{T}x=y$

$\left\{\begin{array}{l}
9x_{1} - 5x_2 + 5x_3=59 \\
5x_{2}+2x_{3}=-33 \\
3x_{3}=-12
\end{array}\right.$

$\left\{\begin{array}{l}
x_{3}=-4 \\
x_{2}=\dfrac{-33 - 2\cdot (-4)}{5} = -5\\
x_{1}=\dfrac{59 + 5 \cdot 5 + 5 \cdot 4}{9} = \dfrac{54}{9} = 6
\end{array}\right.$

Матрица должна быть строго квадратная, невырожденная и вектор коэффициентов соответствующей размерности.

В функции будем приводить матрицу к единичному виду.

Тогда вектор $b$ после преобразований будет равен искомому значению вектора $x$

```python
def mult(A, b):
    n = len(A)

    for row in range(n):
        c = 1 / A[row][row]  # В данном блоке получаем единицу на диагонали в строке с номером row
        for j in range(n):   # В след. блоке будем эту строку домноженную вычитать из остальных, обнуляя все коэффициенты в столбце row
            A[row][j] *= c
        b[row][0] *= c

        for i in range(n):
            if i != row:
                k = A[i][row]
                for j in range(n):
                    A[i][j] -= k * A[row][j]
                b[i][0] -= k * b[row][0]
    return b  
```

```python
A = [[ 2, 1, 3],
     [11, 7, 5],
     [ 9, 8, 4]]

B = [[ 1],
     [-6],
     [-5]]
              
print(*mult(A, B))
```
